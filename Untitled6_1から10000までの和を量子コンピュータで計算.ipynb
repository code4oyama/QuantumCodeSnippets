## 2. 量子コンピュータ用の環境セットアップ

Qiskitのシミュレータ（Aer）を利用して、量子回路の実行環境を準備します。

```python
# シミュレータのバックエンドを取得
from qiskit_aer import AerSimulator
backend = AerSimulator()
print("Qiskitシミュレータの準備ができました。")
```

## 3. 1から10000までの数字を順に足す量子プログラムの作成

実際には量子回路で大規模な加算（1〜10000の和）は現状困難なため、
ここでは量子加算回路の簡単な例と、古典的な計算による合計値も併記します。

```python
# 古典的な計算で1から10000までの和を求める
classic_sum = sum(range(1, 10001))
print(f"古典的な計算による1から10000までの和: {classic_sum}")

# 量子加算回路の簡単な例（2ビット加算）
# ここでは量子加算のイメージのみ示します
qc = QuantumCircuit(4, 2)
# 例: 1 + 1 を量子回路で表現（実際の大規模加算は困難）
qc.x(0)  # 1をセット
qc.x(1)  # 1をセット
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)
qc.measure([2,3], [0,1])
display(qc.draw('mpl'))
```

### 回路図中のq0〜q3, cの意味

- **q0, q1**: 入力となる2つの量子ビット（ここでは1+1の各ビットを表現）
- **q2**: 加算結果（和）を格納する量子ビット
- **q3**: キャリー（繰り上がり）を格納する量子ビット
- **c[0], c[1]**: 量子ビットq2, q3の測定結果を格納する古典ビット

この回路は「1+1」の加算を量子回路で表現したもので、
q2, q3の測定値がそれぞれ和とキャリーを示します。

## 4. 量子プログラムの実行と結果の取得

量子加算回路（例: 1+1）の実行結果を取得します。

```python
# 量子回路の実行
from qiskit_aer import AerSimulator
from qiskit import transpile
# transpileで回路をシミュレータ用に変換
tqc = transpile(qc, backend)
job = backend.run(tqc, shots=1024)
result = job.result()
counts = result.get_counts()
print("量子加算回路の測定結果:", counts)
plot_histogram(counts)
```

- 古典的な計算による1から10000までの和: 50005000
