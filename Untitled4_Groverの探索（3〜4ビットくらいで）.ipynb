# Groverの探索アルゴリズムの実装

このノートブックでは、Groverの量子探索アルゴリズムを用いて、特定のビット列を効率的に探索する回路を実装し、シミュレーション結果を確認します。

---

## Groverのアルゴリズムとは？

- **Groverのアルゴリズム**: 量子計算の代表的な探索アルゴリズムで、N個のデータから特定の1つを√N回の探索で見つけることができる。
- **Oracle**: 解となるビット列にだけ「しるし（位相反転）」をつける量子回路。
- **Diffusion演算子**: 全体の平均から反転する操作で、解の確率を増幅する。
- **反復回数**: √N回程度の繰り返しで最適な確率になる。

---

この実験では、3ビットの探索空間で特定のビット列を効率的に見つける様子を体験できます。

> Groverのアルゴリズムは量子計算の「探索の高速化」の代表例です。

```python

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import GroverOperator
import matplotlib.pyplot as plt
```

## Oracle（マーク状態の位相反転）の作成

探索したいビット列（解）に対して位相反転を行うOracle回路を作成します。

```python

# ────────────── パラメータ ──────────────
n = 3                  # 検索対象の量子ビット数（3ビット → 8通り）
marked_state = '101'   # マークしたい解（ここを変えると別の解を探せる）
iterations = 2        # 最適反復回数 ≈ √(2^n) / √M （M=解の数、ここでは1）

```

```python

# ────────────── Oracle（マークされた状態を位相反転） ──────────────
def phase_oracle(marked_state):
    n = len(marked_state)
    oracle = QuantumCircuit(n)
    for i, bit in enumerate(reversed(marked_state)):
        if bit == '0':
            oracle.x(i)
    oracle.h(n-1)
    if n == 1:
        oracle.z(0)
    else:
        oracle.mcx(list(range(n-1)), n-1)
    oracle.h(n-1)
    for i, bit in enumerate(reversed(marked_state)):
        if bit == '0':
            oracle.x(i)
    return oracle
```

```python

oracle = phase_oracle(marked_state)

# ────────────── Diffusion operator（反転 about average） ──────────────
diffusion = QuantumCircuit(n)
diffusion.h(range(n))
diffusion.x(range(n))
diffusion.h(n-1)
diffusion.mcx(list(range(n-1)), n-1)
diffusion.h(n-1)
diffusion.x(range(n))
diffusion.h(range(n))

```

```python

def run_grover(n, oracle, diffusion, iterations):
    qc = QuantumCircuit(n, n)
    qc.h(range(n))
    qc.barrier()
    for _ in range(iterations):
        qc.compose(oracle, inplace=True)
        qc.compose(diffusion, inplace=True)
        qc.barrier()
    qc.measure(range(n), range(n))
    return qc
```

```python

qc = run_grover(n, oracle, diffusion, iterations)

print(f"Grover回路 (n={n}, marked='{marked_state}', iterations={iterations}):")
display(qc.draw(output='mpl'))
```

```python

# ────────────── シミュレーション実行 ──────────────
simulator = AerSimulator()
job = simulator.run(qc, shots=2048)
result = job.result()
counts = result.get_counts(qc)

print("\n測定結果 (counts):", counts)
plot_histogram(counts)
```
