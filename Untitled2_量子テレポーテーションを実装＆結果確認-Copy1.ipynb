```python



# セル2: 量子テレポーテーション（最新スタイル）
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt

# 3量子ビット + 3古典ビット
qc = QuantumCircuit(3, 3)

# ステップ0: 転送したい状態を q0 に準備（例: |1⟩）
qc.x(0)  # ここを qc.h(0) などに変えて遊べます

# 転送前の状態を表示
print("転送したい状態 (q0):")
sv_before = Statevector.from_instruction(qc)
display(plot_bloch_multivector(sv_before))

# ステップ1: EPRペア作成 (q1 と q2)
qc.h(1)
qc.cx(1, 2)

qc.barrier()

# ステップ2: AliceのBell測定 (q0 と q1)
qc.cx(0, 1)
qc.h(0)

qc.barrier()

# ステップ3: Aliceが測定
qc.measure(0, 0)
qc.measure(1, 1)

# ステップ4: Bobの修正（制御フローで条件分岐）
with qc.if_test((1, 1)):   # 古典ビット1が1なら
    qc.x(2)                # Xゲート（ビットフリップ修正）

with qc.if_test((0, 1)):   # 古典ビット0が1なら
    qc.z(2)                # Zゲート（位相修正）

# Bobの量子ビット(q2)を測定
qc.measure(2, 2)

# 回路図
print("\n量子テレポーテーション回路（最新版）:")
display(qc.draw(output='mpl'))
plt.show()

# シミュレーション
simulator = AerSimulator()
job = simulator.run(qc, shots=4096)
result = job.result()
counts = result.get_counts()

print("\n測定結果:", counts)
plot_histogram(counts)
plt.show()


```
