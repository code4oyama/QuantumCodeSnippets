# 1量子ビットのHゲートと確率の確認

このノートブックでは、量子ビット（qubit）にアダマール（H）ゲートを適用し、量子重ね合わせ状態を作り、その測定結果の確率分布を確認します。

---

## 量子ビット・Hゲート・重ね合わせとは？

- **量子ビット（qubit）**: 0と1の両方の状態を同時に持つことができる量子情報の最小単位です。
- **アダマール（H）ゲート**: 量子ビットを「0と1の重ね合わせ状態」にする基本的な量子ゲートです。
- **重ね合わせ状態**: 量子ビットが「0」と「1」の両方の状態に確率的に存在する状態。
- **測定**: 量子ビットの状態を観測し、0または1のいずれかの値として得る操作。

---

この実験では、1つの量子ビットにHゲートをかけて重ね合わせ状態を作り、何度も測定することで、理論上「0」と「1」がほぼ50%ずつ出現することを確認します。

> 量子の「重ね合わせ」や「確率的な測定結果」は、量子計算の本質的な特徴です。

## 量子回路の作成とHゲートの適用

1量子ビット回路を作成し、アダマール（H）ゲートを適用して重ね合わせ状態を作ります。


## 必要なライブラリのインポート


| モジュール/関数名                | 説明                                                         |
|:----------------------------------|:-------------------------------------------------------------|
| QuantumCircuit                    | 量子回路を作成・操作するためのQiskitの基本クラス              |
| AerSimulator                      | Qiskitのシミュレータ（量子回路を計算機上で実行するためのもの）|
| plot_histogram                    | 測定結果などのヒストグラムを描画する関数                     |
| matplotlib.pyplot as plt          | グラフ描画用のPython標準ライブラリmatplotlibのインターフェース|

```python

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

| コード                                      | 説明                                                                 |
|:--------------------------------------------|:---------------------------------------------------------------------|
| qc = QuantumCircuit(1, 1)                   | 1量子ビットと1古典ビットの量子回路を作成します。                    |
| qc.h(0)                                     | 量子ビット0にアダマール（H）ゲートを適用し、重ね合わせ状態にします。|
| qc.measure(0, 0)                            | 量子ビット0を古典ビット0に測定します。                              |
| print("回路図:")                            | 回路図を表示する前にラベルを出力します。                            |
| display(qc.draw(output='mpl'))              | 回路図をmatplotlibで可視化します。                                   |

```python

# 量子回路を作成（1量子ビット + 1古典ビット）
qc = QuantumCircuit(1, 1)

# アダマールゲートをかける → 重ね合わせ状態に
qc.h(0)

# 測定する
qc.measure(0, 0)

# 回路を可視化（どんな感じか見てみよう）
print("回路図:")
display(qc.draw(output='mpl'))

```

## 回路の可視化とシミュレーション実行

作成した回路を描画し、シミュレータで多数回実行して測定結果の確率分布を確認します。


| コード                              | 説明                                                                 |
|:------------------------------------|:---------------------------------------------------------------------|
| simulator = AerSimulator()          | Qiskitのシミュレータを準備します。                                   |
| job = simulator.run(qc, shots=1024) | 量子回路qcを1024回実行し、ジョブを作成します。                      |
| result = job.result()               | 実行結果（Resultオブジェクト）を取得します。                        |
| counts = result.get_counts(qc)      | 測定結果（各ビット列の出現回数）を取得します。                      |
| print("\n測定結果（counts）:", counts) | 測定結果（counts）をコンソールに表示します。                        |
| plot_histogram(counts)              | 測定結果をヒストグラムで可視化します。                               |

```python

# シミュレータを準備
simulator = AerSimulator()

# 回路を1024回実行
job = simulator.run(qc, shots=1024)

# 結果を取得
result = job.result()
counts = result.get_counts(qc)

# 結果を表示
print("\n測定結果（counts）:", counts)

# ヒストグラムで可視化
plot_histogram(counts)
```
