## Bell状態・もつれ・CNOTゲートとは？

- **Bell状態**: 2つの量子ビットが「もつれ」た状態で、どちらか一方を測定するともう一方も必ず同じ結果になる量子状態。
- **もつれ（エンタングルメント）**: 複数の量子ビットが独立でなく、測定結果が強く相関する量子現象。
- **CNOTゲート**: 1つ目の量子ビットが1のときだけ2つ目の量子ビットを反転させる量子ゲート。もつれ状態の生成に使われる。
- **測定**: 量子ビットの状態を観測し、0または1のいずれかの値として得る操作。

---

この実験では、HゲートとCNOTゲートを使ってBell状態を作り、何度も測定することで「00」と「11」だけが出現し、完全な相関があることを確認します。

> 量子もつれは量子計算・量子通信の基礎となる重要な現象です。

## Bell状態を作る量子回路の作成

2量子ビット回路を作成し、HゲートとCNOTゲートでもつれ状態（Bell状態）を生成します。

```python


from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# 2量子ビット + 2古典ビットの回路を作成
qc = QuantumCircuit(2, 2)

# ステップ1: 最初の量子ビットにHゲート（重ね合わせにする）
qc.h(0)

# ステップ2: CNOTゲートで2番目の量子ビットを制御（もつれを作る）
qc.cx(0, 1)   # control=0, target=1

# 測定（両方の量子ビットを測定）
qc.measure([0, 1], [0, 1])

# 回路図を表示
print("回路図:")
display(qc.draw(output='mpl'))
plt.show()

# シミュレータを準備
simulator = AerSimulator()

# 1024回実行（shotsを増やすとより綺麗な分布になる）
job = simulator.run(qc, shots=2048)

# 結果を取得
result = job.result()
counts = result.get_counts(qc)

# 結果を表示
print("\n測定結果（counts）:", counts)

# ヒストグラムで可視化
plot_histogram(counts)
plt.show()

```
